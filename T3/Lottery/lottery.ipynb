Task 1: Probability of Winning the Grand Prize
    Winning the grand prize in a state lottery, by matching six numbers drawn from a 
    one to fifty-one number is considered a discrete probability. Due to finite number of possible 
    outcomes, and each outcome is non-continous.

    The code in task 1 would need to:
    - Calculate probability of winning grand prize for a new game
    - Given that a group of 6 is selected from {1,2,3, ..., 51}


```python
# Import libraries
import math
import matplotlib.pyplot as plt

def probability_of_winning():
    # Total number of balls in the set
    total_balls = 51

    # Number of balls to draw
    balls_to_draw = 6

    # Number of ways to win (matching all 6 numbers)
    ways_to_win = 1  # There is only one winning combination

    # Total number of possible combinations
    total_combinations = math.comb(total_balls, balls_to_draw)

    # Calculate the probability
    probability = ways_to_win / total_combinations

    return probability

# Call the function to get the probability of winning
winning_probability = probability_of_winning()
losing_probability = 1 - winning_probability

print(f"The probability of winning the grand prize is approximately {winning_probability:.10f}")

# Create a bar chart to visualize winning and losing probabilities
categories = ['Winning', 'Losing']
probabilities = [winning_probability, losing_probability]

plt.figure(figsize=(8, 6))
plt.bar(categories, probabilities, color=['green', 'red'])
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Winning and Losing Probabilities')
plt.ylim(0, 1)  # Set the y-axis limit to [0, 1]
plt.show()

```

Task 2: Probability of Alexandra Winning
- Alexandra purchase 1 ticket every 24 hours for 365 days for 80 years
- A ticket costs $5 
- Numbers are selected independently from one ticket to the next
- Calculate probability of winning a lottery at least 1 time
- Calculate amount of money spent trying

```python
# Import Libraries
from math import comb
import matplotlib.pyplot as plt

# Create function to find the probability of winning 
def calculate_probability_of_winning():
    # Total number of possible combinations
    total_combinations = comb(51, 6)
    
    # Number of ways to win (1 combination out of total)
    ways_to_win = 1
    
    # Probability of winning
    probability = ways_to_win / total_combinations
    
    return probability

# Create function to simulate lottery
def simulate_lottery(years, tickets_per_day, cost_per_ticket):
    # Calculate the probability of winning
    probability_of_winning = calculate_probability_of_winning()
    
    # Calculate the probability of not winning
    probability_of_not_winning = 1 - probability_of_winning
    
    # Calculate the total number of tickets purchased
    total_tickets_purchased = years * 365 * tickets_per_day
    
    # Calculate the total cost
    total_cost = total_tickets_purchased * cost_per_ticket
    
    # Calculate the probability of not winning for all purchased tickets
    probability_of_not_winning_all = (probability_of_not_winning) ** total_tickets_purchased
    
    # Calculate the probability of winning at least once
    probability_of_winning_at_least_once = 1 - probability_of_not_winning_all
    
    return probability_of_winning_at_least_once, total_cost

# Define the parameters
years = 80
tickets_per_day = 1
cost_per_ticket = 5

# Simulate Alexandra's lottery purchases
probability, total_cost = simulate_lottery(years, tickets_per_day, cost_per_ticket)

# Create a bar chart to visualize the probability
plt.figure(figsize=(10, 5))
plt.bar(["Probability of Winning at Least Once", "Total Cost"], [probability, total_cost], color=['blue', 'green'])
plt.ylabel("Value")
plt.title("Probability of Winning the Lottery and Total Cost")
plt.ylim(0, 1.1 * max(probability, total_cost))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Print out results
print(f"The probability of Alexandra winning the lottery at least once in {years} years is approximately {probability:.10f}")
print(f"The amount of money Alexandra spends trying is ${total_cost:.2f}")


```

Task 3: Amir and Friends
- Amir decides to increase chances of winning with N friends buying tickets independently
- Minimum value of N (Nmin epsilon) so at least one winner among the group is greater than or equal to epsilon
- Calculate value of Nmin(epsilon) for epsilon 1e-5, 1e-3, 0.1, 0.5

Using the complimentary probbility formula, we could derive a formula that would show
the number of friends needed to be above or equal the threshold for epsilon.
Complemtary Formula in this example would state that the probability of not winning is 1-P.
Due to each event being independent, If N friedns each buy a ticket, the probability that none of them wins is (1-P)^N.
To find at least one friend wins, we take the complement which is 1- (1-P)^N.
This will give us the a formula to find the number of friends such that it is greater or equal to the thresholf.
ε <= 1 - (1-P)^N
1 - ε >= (1-P)^N
ln(1 - ε) >= N * ln(1-P)
ln(1 - ε) / (ln(1-P)) >= N


```python
# Import libraries
import math
from math import comb

# Create function to find the probability of winning 
def calculate_probability_of_winning():
    # Total number of possible combinations
    total_combinations = comb(51, 6)
    
    # Number of ways to win (1 combination out of total)
    ways_to_win = 1
    
    # Probability of winning
    probability = ways_to_win / total_combinations
    
    return probability

# Create a function that would find the Nmin(epsilon)
# Using the complementary probability
def calculate_N_min_for_epsilon(epsilon):
    # Calculate the probability of winning with one ticket purchase
    P = calculate_probability_of_winning()
    
    # Calculate N_min using the formula
    N_min = math.ceil(math.log(1 - epsilon) / math.log(1 - P))
    
    return N_min

# Define a list of epsilon values for which you want to calculate N_min
epsilon_values = [1e-5, 1e-3, 0.1, 0.5]

# Calculate and print N_min for each epsilon value
for epsilon in epsilon_values:
    N_min = calculate_N_min_for_epsilon(epsilon)
    print(f"For ε = {epsilon}, N_min = {N_min} ")


```

Task 4: Amir and Friends, Incresing Their Win Probability
- If Amir gathered enough friends to win with 0.5 probability using Task 3 strategy
- Determine a way they can increase their win probability above .5 without adding more people

```python
# Import libraries
import math
from math import comb
import matplotlib.pyplot as plt

# Calculate probabilit of wining
def calculate_probability_of_winning():
    # Total number of possible combinations
    total_combinations = comb(51, 6)
    
    # Number of ways to win (1 combination out of total)
    ways_to_win = 1
    
    # Probability of winning
    probability = ways_to_win / total_combinations
    
    return probability
# Create a function to find the number of minimun friends per threshold
def calculate_N_min_for_epsilon(epsilon):
    # Calculate the probability of winning with one ticket purchase
    P = calculate_probability_of_winning()
    
    # Calculate N_min using the formula
    N_min = math.ceil(math.log(1 - epsilon) / math.log(1 - P))
    
    return N_min

# Calculate the number of tickets with the number of friend with the 0.5 threshold
def calculate_combined_win_probability(N, num_tickets):
    # Calculate the probability of winning with a group strategy
    P_single = calculate_probability_of_winning()
    P_combined = 1 - (1 - P_single) ** (N * num_tickets)
    
    return P_combined

# Define the epsilon value (desired probability threshold)
epsilon = 0.5

# Calculate N_min for the initial strategy
N_min = calculate_N_min_for_epsilon(epsilon)

# Initialize lists to store data for visualization
num_tickets_values = []
win_probability_values = []

# Calculate and store probabilities for different numbers of tickets purchased together
for num_tickets in range(1, 11):
    num_tickets_values.append(num_tickets)
    win_probability = calculate_combined_win_probability(N_min, num_tickets)
    win_probability_values.append(win_probability)

# Create a visualization that shows the win probability over the number purchased
plt.figure(figsize=(10, 6))
plt.plot(num_tickets_values, win_probability_values, marker='o')
plt.title('Win Probability vs. Number of Tickets Purchased Together')
plt.xlabel('Number of Tickets Purchased Together')
plt.ylabel('Win Probability')
plt.grid(True)
plt.show()

# Print the results
print(f"With the initial strategy (each friend buys 1 ticket),")
print(f"N_min = {N_min} friends are needed for ε = {epsilon}.")
print(f"Probability of winning with the initial strategy: {win_probability_values[0]:.10f}")

```

Task 5: Consolation Prize
- If exactily 3 of the 6 numbers match any of the state's, you win a smaller prize
- Calculate the probability of winning the consolation prize
- Calculate the expected value of the consolation prize

The addition of a consolation prize could likely increase the sales of lottery tickets. However, 
this also depends on the amount of the prize, ticket prices, and player preferences. The idea of a 
consolation prize could attract numerous players by encouraging them that they would only need three out of six balls to win, however this also depends on marketing and promotional efforts .

```python
import math
from math import comb
import matplotlib.pyplot as plt

def calculate_probability_of_winning():
    # Total number of possible combinations
    total_combinations = comb(51, 6)
    
    # Number of ways to win the consolation prize (3 out of 6 numbers)
    ways_to_win = comb(6, 3) * comb(45, 3)
    
    # Probability of winning the consolation prize
    probability = ways_to_win / total_combinations
    
    return probability

def calculate_expected_value(probability, prize_amount):
    # Expected value = Probability of winning * Prize amount
    expected_value = probability * prize_amount
    
    return expected_value

# Define the prize amount for the consolation prize
prize_amount = 1200000000  

# Calculate the probability of winning the consolation prize
probability = calculate_probability_of_winning()

# Calculate the expected value of the consolation prize
expected_value = calculate_expected_value(probability, prize_amount)

# Create a bar chart to visualize the prize amount and expected value
labels = ["Prize Amount", "Expected Value"]
values = [prize_amount, expected_value]
colors = ["blue", "green"]

plt.figure(figsize=(8, 6))
plt.bar(labels, values, color=colors)
plt.title("Prize Amount vs. Expected Value")
plt.ylabel("Amount ($)")
plt.grid(axis="y")
plt.show()

print(f"The probability of winning the consolation prize is approximately {probability:.10f}")
print(f"The expected value of the consolation prize is ${expected_value:.2f}")
# Determine if the consolation prize is likely to increase sales
if expected_value > 0:
    print("The consolation prize is likely to increase sales.")
else:
    print("The consolation prize may not significantly impact sales.")
```
