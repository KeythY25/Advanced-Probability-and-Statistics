1. Visualization of how many houses are in the longitude axis
Scatter plot of where houses are located in longitude and latitude


```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Import csv file
csv_file = 'housing.csv'
data = pd.read_csv(csv_file)

# Create histogram for the frequency of longitude
plt.figure(figsize=(10, 5))
plt.hist(data['longitude'], bins=30, color='skyblue')
plt.title('Histogram of Longitude')
plt.xlabel('Longitude')
plt.ylabel('Frequency')
plt.show()

# Create scatter plot to show longitude and latitude
plt.figure(figsize=(10, 5))
plt.scatter(data['longitude'], data['latitude'], alpha=0.2, c='blue')
plt.title('Scatter Plot of Latitude vs Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

```

2. Total Rooms and Median Housing Age

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import CSV file
csv_file = 'housing.csv'
data = pd.read_csv(csv_file)


# Create a scatter plot for total rooms and Housing Median Age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='housing_median_age', y='total_rooms', data=data, alpha=0.5)
plt.title('Scatter Plot of Housing Median Age vs. Total Rooms')
plt.xlabel('Housing Median Age')
plt.ylabel('Total Rooms')
plt.grid(True)

# Create Age intervals
intervals = [0, 10, 20, 30, 40, 50, 60]

# Group data into 10 year intervals
data['Age Interval'] = pd.cut(data['housing_median_age'], bins=intervals, labels=[f'{i}-{i+10}' for i in intervals[:-1]])
plt.figure(figsize=(12, 6))
sns.violinplot(x='Age Interval', y='total_rooms', data=data, palette='viridis', inner='quartile')
plt.title('Violin Plot of Total Rooms by Housing Median Age (10-Year Intervals)')
plt.xlabel('Age Interval')
plt.ylabel('Total Rooms')

# Create median labels
medians = data.groupby('Age Interval')['total_rooms'].median().values
median_labels = [str(int(median)) for median in medians]

# Add Median labels in the violin chart
for i, median in enumerate(medians):
    plt.text(i, median, median_labels[i], horizontalalignment='center', size='medium', color='red', weight='semibold')

plt.show()
```

3. Total Bedrooms vs Total Rooms

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import CSV file
csv_file = 'housing.csv'
data = pd.read_csv(csv_file)

# Create histograms for 'total_rooms' and 'total_bedrooms'
plt.figure(figsize=(12, 6))
plt.hist(data['total_rooms'], bins=30, color='skyblue', alpha=0.7, label='Total Rooms')
plt.hist(data['total_bedrooms'], bins=30, color='orange', alpha=0.7, label='Total Bedrooms')
plt.title('Histograms of Total Rooms and Total Bedrooms')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.show()

# Create a pair plot for 'total_rooms' and 'total_bedrooms'
sns.set(style='whitegrid')
pair_plot = sns.pairplot(data[['total_rooms', 'total_bedrooms']], diag_kind='kde', height=3)
pair_plot.fig.suptitle('Pair Plot of Total Rooms and Total Bedrooms', y=1.02)
plt.show()
```

4. Population vs. Household

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import CSV file
csv_file = 'housing.csv'
data = pd.read_csv(csv_file)

# Create a scatter plot for 'Population' and 'Households'
plt.figure(figsize=(8, 6))
plt.scatter(data['population'], data['households'], alpha=0.5, color='blue')
plt.title('Scatter Plot of Population vs. Households')
plt.xlabel('Population')
plt.ylabel('Households')
plt.grid(True)
plt.show()

# Create a dataFrame
df = data[['population', 'households']]

# Create a pair plot to visualize the relationships
sns.set(style='whitegrid')
pair_plot = sns.pairplot(df, height=3)
pair_plot.fig.suptitle('Pair Plot of Population and Households', y=1.02)
plt.show()

# Sort the value by population

data = data.sort_values(by='population')

# Create a line graph for 'Population' and 'Households'
plt.figure(figsize=(10, 6))
plt.plot(data['population'], data['households'], color='blue', marker='o')
plt.title('Line Graph of Population vs. Households')
plt.xlabel('Population')
plt.ylabel('Households')
plt.grid(True)
plt.show()
```

6. Median Income, Median House Value and Median House Age

```python
# Import Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Import CSV file
csv_file = 'housing.csv'
data = pd.read_csv(csv_file)

# Create a DataFrame with relevant columns
df = data[['median_income', 'housing_median_age', 'median_house_value']]

# Create a scatter plot matrix
sns.set(style='whitegrid')
scatter_matrix = sns.pairplot(df, height=3)
scatter_matrix.fig.suptitle('Scatter Plot Matrix of Median Income, House Median Age, and Median House Value', y=1.02)
plt.show()

# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Median Income, House Median Age, and Median House Value')
plt.show()

```

7. Median House Value and Ocean Proximity

```python
# import Libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Import CSV file
csv_file = 'housing.csv'
data = pd.read_csv(csv_file)

# Create a box plot for 'Median House Value' by 'Ocean Proximity'
plt.figure(figsize=(10, 6))
box_plot = sns.boxplot(x='ocean_proximity', y='median_house_value', data=data, palette='viridis')
plt.title('Box Plot of Median House Value by Ocean Proximity')
plt.xlabel('Ocean Proximity')
plt.ylabel('Median House Value')
plt.xticks(rotation=45)

# Create median labels
medians = data.groupby('ocean_proximity')['median_house_value'].median()
vertical_offset = data['median_house_value'].median() * 0.05  
# Add median labels in the graph
for xtick in box_plot.get_xticks():
    box_plot.text(xtick, medians[xtick] + vertical_offset, f'{medians[xtick]:.2f}', 
                  horizontalalignment='center', size='medium', color='r', weight='semibold')

plt.show()

# Create a histogram for 'Ocean Proximity'
plt.figure(figsize=(8, 5))
sns.histplot(data=data, x='ocean_proximity', discrete=True, palette='viridis')
plt.title('Histogram of Ocean Proximity')
plt.xlabel('Ocean Proximity')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Create a counts label and add into the histogram
counts = data['ocean_proximity'].value_counts()
for i, count in enumerate(counts):
    plt.text(i, count, count, ha='center', va='bottom', fontsize=12)
plt.show()
```

8. Population vs. Ocean Proximity


```python
# Import Libraries 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Import CSV file
csv_file = 'housing.csv'
data = pd.read_csv(csv_file)

# Create a violin plot for 'Population' by 'Ocean Proximity'
plt.figure(figsize=(12, 6))
violin_plot = sns.violinplot(x='ocean_proximity', y='population', data=data, palette='viridis')
plt.title('Violin Plot of Population by Ocean Proximity')
plt.xlabel('Ocean Proximity')
plt.ylabel('Population')
plt.xticks(rotation=45)

# Create median labels and add into the graph
medians = data.groupby('ocean_proximity')['population'].median()
vertical_offset = data['population'].median() * 0.05  # Adjust the offset as needed
for i, median in enumerate(medians):
    violin_plot.text(i, median + vertical_offset, f'{median:.2f}', ha='center', va='bottom', fontsize=10)

plt.show()

# Create custom colors for the pie chart
custom_colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#c2c2f0']

# Calculate the counts of each 'Ocean Proximity' category
ocean_proximity_counts = data['ocean_proximity'].value_counts()

# Create a pie chart with the custom color and starting from the 90 degree 
plt.figure(figsize=(8, 8))
plt.pie(ocean_proximity_counts, labels=ocean_proximity_counts.index, autopct='%1.1f%%', startangle=90, colors=custom_colors, counterclock=False)
plt.title('Pie Chart of Ocean Proximity Distribution')
plt.axis('equal')

plt.show()
```
